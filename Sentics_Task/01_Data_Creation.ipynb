```python
import pandas as pd
import numpy as np
```

### **Here I have made various assumptions,**
* Considered 5 cameras/Sensors named : A , B , C , D , E
* Random values of X and Y co-ordinates are generated. 
* Three Uinque ID's considered 
    * 126 : Object 1
    * 348 : Object 2 
    * 0 : Unknown Object
* Ids are the Local Id and unique Id is the Global Id


```python
np.random.seed(2024)
timestamps = np.repeat(np.arange(1, 5), 5)
x_positions = np.random.uniform(0, 100, 20)
y_positions = np.random.uniform(0, 100, 20)
unique_ids = np.random.choice([0, 126, 348], 20)
sensor_ids = np.tile(['A', 'B', 'C', 'D', 'E'], 4)
ids = np.arange(1, 21)
```

```python
created_data = pd.DataFrame({
     'timestamp': timestamps,
    'id': ids,
    'x_position': x_positions,
     'y_position': y_positions,
    'unique_id': unique_ids,
    'sensor_id': sensor_ids})
```

```python
created_data.to_csv('created_data.csv', index=False)
```
